```python
# import libraries
```

```python
import pandas as pd
import numpy as np
```

```python
data_path ="/content/db27810d5f034f92b924bbbc027863c4.csv"
dataset = pd.read_csv(data_path)
```

```python
# split data into feature and target
```

```python
feature = dataset.iloc[: , :-1].values
target = dataset.iloc[: ,-1].values
```

```python
# split to train and test
```

```python
from sklearn.model_selection import train_test_split
feature_train, feature_test, target_train, target_test = train_test_split(feature,target,test_size=.25 )
```

```python
# draw relation
```

```python
import matplotlib.pyplot as plt

plt.scatter(feature_train,target_train,color="red")
```

```python
import seaborn as sns
# sns.regplot(x=feature_train,y=target_train,ci=None,color="red")
# plt.show()
#
print(sns.regplot(x=feature_train,y=target_train,color="red")
)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
model = LinearRegression()
model.fit(feature_train , target_train)
```

```python
slope_m = model.coef_[0]
```

```python
intercept_c = model.intercept_
```

```python
target_pred = model.predict(feature_test)
```

```python
model.score(feature_test,target_test)
```

```python
target = (slope_m)*feature_train + (intercept_c)
```

```python
plt.plot(feature_train,target,color="blue")
plt.scatter(feature_train,target_train,color="red")
```
